## Loop

```python
apiKey = 'R/Tc95/XWkozetyTxFacnZ+YoC7U4O+ZMbPIVc9EJTYH3m3lm/n75YMNLbozV1gLHzF/vAxZzrURxwIB+Dz7dg=='
url = 'http://apis.data.go.kr/1130000/publicYmList/publicYmListApi'
dictPublicYm = {}

for year in range(2001, 2023):
    params ={'serviceKey' : apiKey, 'pageNo' : '1', 'numOfRows' : '100', 'jobSeCode' : '0001', \
             'presentnYear' : str(year)}

    out = f'../data/raw/api/publicYmListApi{year}.xml'
    response = requests.get(url, params=params)
    # check response
    # print(response.content)
    with open(out, 'wb') as outFile:
        outFile.write(response.content)
    tree = ET.parse(out)
    root = tree.getroot()
    for ym in root.iter('othbcYm'):
        print(ym.text)
    dictPublicYm[str(year)] = ym.text
```

```python
dfYm = pd.DataFrame.from_dict(dictPublicYm, orient='index', columns=['publicYm'])
dfYm
```

```python
with open('../data/lrgBsnsGrps/publicYm/publicYm.pkl', 'wb') as dump:
    pickle.dump(dictPublicYm, dump)
```
