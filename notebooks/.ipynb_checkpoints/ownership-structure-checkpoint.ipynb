# 소유지배구조 특성

We classify ownership structure as 1. unified holding company structrue 2. holding company structrue 3. holding company 4. loop structure 5. etc.

```python
def ownStrucType(row):
    if (row['hldStruc'] >= 1) and (row['nHld'] < 2):
        return 'unifiedHldStruc'
    elif (row['hldStruc'] >= 1):
        return 'hldStruc'
    elif (row['hldExist'] >= 1):
        return 'hld'
    else:
        return 'etc.'
```

```python
dfChaebol21['ownStrucType'] = dfChaebol21.apply(ownStrucType, axis=1)
dfChaebol21
```

```python
lOwnershipInput = []
for grp in dfChaebol21['unityGrupNm'].unique():
    df = pd.read_excel(f"../data/almeida/2021/in/{grp}/{grp}2021raw.xlsx")
    lOwnershipInput.append(df)
dfOwnershipInput = pd.concat(lOwnershipInput)
print(len(dfOwnershipInput))
dfOwnershipInput.head()
```

```python
dfVr = pd.DataFrame(dfOwnershipInput[dfOwnershipInput['type'] != 99.].groupby("firmid")['own2'].sum()).reset_index()
dfVr.columns = ['firmid', 'vr']
dfVr.head()
```

```python
dfOwn.head()
```

```python
dfChaebol21.head()
```

```python
dfCcVr = dfOwn.merge(dfChaebol21[['unityGrupNm', 'loopExist', 'ownStrucType']], how='left',\
                     left_on='grpname2', right_on='unityGrupNm').drop(columns=['unityGrupNm']).merge(dfVr,
                                                                                                     on='firmid', how='left')
dfCcVr.head()
```

```python
dfFam1 = pd.DataFrame(dfOwnershipInput[dfOwnershipInput['type'] == 0]\
                      .groupby("firmid")['own'].sum()).reset_index()
dfFam1.columns = ['firmid', 'fam1']
dfFam1.head()
```

```python
dfCcVrFam1 = dfCcVr.merge(dfFam1, on='firmid', how='left')
dfCcVrFam1.head()
```

```python
dfAffil = pd.DataFrame(dfOwnershipInput[dfOwnershipInput['type'] == 1].groupby("firmid")['own'].sum()).reset_index()
dfAffil.columns = ['firmid', 'affil']
dfAffil.head()
```

```python
dfCcVrFam1Affil = dfCcVrFam1.merge(dfAffil, on='firmid', how='left')
dfCcVrFam1Affil.head()
```

```python
dfCcVrFam1Affil['wedge1'] = (dfCcVrFam1Affil['lastcontrol'] / 100) - dfCcVrFam1Affil['ultown']
dfCcVrFam1Affil['wedge2'] = dfCcVrFam1Affil['vr'] - dfCcVrFam1Affil['ultown']
```
