```python
dfTransactionsRankVr2['실질지분율 차이'] = dfTransactionsRankVr2['매출회사 실질지분율'] - dfTransactionsRankVr2['매입회사 실질지분율']
```

```python
dfTransactionsRankVr2.head(50)
```

```python
def diffType(row):
    if pd.isna(row['실질지분율 차이']):
        return -1
    elif row['실질지분율 차이'] < 0:
        return 0
    elif row['실질지분율 차이'] < 0.03:
        return 1
    else:
        return 2
```

```python
dfTransactionsRankVr2['diffType'] = dfTransactionsRankVr2.apply(diffType, axis=1) 
```
