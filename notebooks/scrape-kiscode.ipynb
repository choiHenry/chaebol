```python
df_1 = pd.read_excel(f'../data/cmpnySumry/cmpnySumry{year+1}Eng.xlsx')
df_1[['jurirno', 'rcmgCode']]
```

```python
df = df.merge(df_1[['jurirno', 'rcmgCode']], on='jurirno', how='left')
```

```python
dfTrgt = df[df['rcmgCode'].isna()]
dfTrgt
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)

grsrch_url = 'https://www.kisline.com/cm/CM0100M00GE00.nice'
username = 'wonbok'
password = 'gkrtkrhk5034!'
driver.get(grsrch_url)
driver.find_element_by_class_name('btn_close_layer').click()
driver.find_element_by_id('lgnuid').send_keys(username)
driver.find_element_by_id('tmp_lgnupassword').click()
driver.find_element_by_id('lgnupassword').send_keys(password)
driver.find_element_by_class_name('btn_log_in').click()
driver.implicitly_wait(3)
sleep(1)

dKiscode = dict()
inputElem = driver.find_element_by_id('q')

for jurirno in dfTrgt['jurirno']:
    
    if jurirno in dKiscode.keys():
        continue
        
    inputElem.send_keys(jurirno)
    driver.find_element_by_id('searchView').click()
    driver.implicitly_wait(3)
    target_row = driver.find_element_by_xpath('//*[@id="eprTable"]/tbody/tr/td[2]')
    kiscode = target_row.get_attribute('data-kiscode')
    dKiscode[jurirno] = kiscode
    print(f"{jurirno}: {kiscode}")
    sleep(3)
    inputElem = driver.find_element_by_id('q')
    inputElem.clear()

sleep(3)
driver.find_element_by_link_text('로그아웃').click()
sleep(1)
driver.quit()
```
